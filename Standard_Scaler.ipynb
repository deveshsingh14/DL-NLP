```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential #ANN
from tensorflow.keras.layers import Dense #Layer
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel("/content/fake_reg.xlsx")
```

```python
df.sample(5)
```

```python
df.info()
```

```python
x = df.iloc[:,1:].values
y = df.iloc[:,0].values
```

```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split (x,y,test_size=0.3,random_state=1)
```

```python
sc = StandardScaler()

xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
# step 1 Initialization Model
ann = Sequential()

#step 2 Add Layer (Unit = No. of neurons)
ann.add(Dense(units=4,activation="relu")) # i/p layer
ann.add(Dense(units=1))# o/p layer # no activation function in o/p layer in
#regressionproblem

#step 3 EStablish connection between layers
ann.compile(optimizer="adam",loss="mse")#no meteic parameter in regression

#step 4 Train Model
ann.fit(xtrain,ytrain,batch_size=20,epochs=250)
```

```python
# step 5 : prediction
ypred=ann.predict(xtest)

print(f"MSE:{mean_squared_error(ytest,ypred)}")
print(f"RMSE:{np.sqrt(mean_squared_error(ytest,ypred))}")
```
